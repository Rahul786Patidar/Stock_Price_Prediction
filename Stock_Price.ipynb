```python
features = ['Open', 'High', 'Low', 'Close', 'Volume']

# Create a figure and subplots
fig, axes = plt.subplots(1, len(features), figsize=(20, 10))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Plot the distribution of each feature
for i, col in enumerate(features):
    sns.histplot(data[col], ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel('Values')
    axes[i].set_ylabel('Density')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Building the regression model
from sklearn.model_selection import train_test_split

#for preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

#for model evaluation
from sklearn.metrics import mean_squared_error as  mse
from sklearn.metrics import r2_score
```

```python
# Split the data into train test split
X = np.array(data.index).reshape(-1,1)
Y = data['Close']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
#feature scaling
scaler = StandardScaler().fit(X_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#creating a linear model
lm = LinearRegression()
lm.fit(X_train,Y_train)
```

```python
import plotly.graph_objs as go

# Assuming you have defined layout somewhere in your code
layout = go.Layout(
    title='Actual vs Predicted Values for Train Dataset',
    xaxis=dict(title='Day'),
    yaxis=dict(title='Values'),
)

# Assuming lm is your trained model
trace0 = go.Scatter(
    x=X_train.T[0],
    y=Y_train,
    mode='markers',
    name='Actual'
)

trace1 = go.Scatter(
    x=X_train.T[0],
    y=lm.predict(X_train).T,
    mode='lines',
    name='Predicted'
)

data = [trace0, trace1]

plot2 = go.Figure(data=data, layout=layout)

# Now you can show or save the plot
plot2.show()
```

```python
#calculate scores for model evaluation
scores = f'''
{'Metrics'.ljust(10)}{'Train'.center(20)}{'Test'.center(20)}
{'r2_score'.ljust(10)}{r2_score(Y_train,lm.predict(X_train))}\t{r2_score(Y_test,lm.predict(X_test))}
{'MSE'.ljust(10)}{mse(Y_train,lm.predict(X_train))}\t{mse(Y_test,lm.predict(X_test))}
'''
print(scores)
# Apply center() to strings
```
